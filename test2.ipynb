1.首先要知道的是np操作多维数组很厉害，大大减少运算时间和for循环的使用

3.数组的索引，轴0作为行，轴1作为列


但是在用np.sum时是返回一样的轴，就是反的相加，np.mean()也一样

```python
a = np.array([[1,2,2],
              [2,3,4],
              [3,4,5]])
b = np.sum(a,axis=0)  #结果是行相加
print(b)
```

4.最近也在看吴恩达的深度学习，吴老师提醒不要创建秩为1的数组，不然会有很多BUG

```python
a = np.ones(3)  #少用！！
b = np.ones((3,1))
print(a)
print(b)
```

5.多条件的布尔索引不能用and和or关键之，只能用&，|，~

```python
a = np.array([[1,2,3],
            [2,3,4],
            [3,4,5]])
s = (a>1) & (a<3)
b = a[s]
print(b)
```

6.np的广播功能是非常强大，可以广播函数，可以广播两个不同形状的数组

```python
a = np.array([[1,2,2],
              [2,3,4],
              [3,4,5]])
b = np.array([[1],
             [2],
             [3]])
c = np.log(a)  #a中每个元素取对数
print(c)

d = a + b  #不同形状广播一般是把形状小的变成形状大的，比如这里单列b数组被复制成3列，然后再和a中每个元素相加
print(d)
```

7.内置函数ceil向上取整，floor向下取整

```python
a = np.ceil(np.array([1.5]))
b = np.floor(np.array([1.5]))
print(a,b)
```

1.填充边界

```python
Z = np.ones((5,5))
Z = np.pad(Z, pad_width=1, mode='constant', constant_values=0)
print(Z)
```

2.查看指定第几元素的索引

```python
print(np.unravel_index(100,(6,7,8)))   #形状为(6,7,8)数组第100个元素索引
```

3.归一化

```python
Z = np.arange(9)
Z = Z.reshape((3,3))
print(Z)
Zmax, Zmin = Z.max(), Z.min()
Z = (Z - Zmin)/(Zmax - Zmin)
print(Z)   #这里是默认按行做归一化
```
