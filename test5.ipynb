```python
mnames = ['movie_id', 'title', 'genres']    #多类别情况
movies = pd.read_table('pydata/datasets/movielens/movies.dat', sep='::',header=None, names=mnames)
movies.head()
```

```python
all_genres = []
for x in movies.genres:
    all_genres.extend(x.split('|'))
genres = pd.unique(all_genres)
genres
```

```python
zero_matrix = np.zeros((len(movies.genres),len(genres)))
dummies = pd.DataFrame(zero_matrix, columns=genres)
dummies
```

```python
for i,gen in enumerate(movies.genres):
    index = dummies.columns.get_indexer(gen.split('|'))    #生成相同列名的列索引列表
    dummies.iloc[i,index] = 1  
dummies.head()
```

```python
movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.iloc[0]  #第一行数据
```

```python
import re
```

```python
text = "foo    bar\t baz  \tqux"
re.split('\s+',text)
```

```python
regex = re.compile('\s+')
regex.split(text)
```

```python
regex.findall(text)
```

```python
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com
"""
pattern = r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'

regex = re.compile(pattern,flags=re.IGNORECASE)
regex.findall(text)
```

```python
b = regex.search(text)
text[b.start():b.end()]
```

```python
print(regex.sub('EMAIL',text))
```

```python
pattern = r'([A-Z0-9._%]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})'
regex = re.compile(pattern,flags=re.IGNORECASE)
regex.findall(text)
```
